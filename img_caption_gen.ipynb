```python
# Webpage python
import gradio as gr
# HuggingFace Model (Salesforce) BLIP
from transformers import AutoProcessor, BlipForConditionalGeneration
# Image uploder to Python env
from PIL import Image
# Useful for converting Image to nd-array
import numpy as np
```

```python
# Using Pre-defined model
processor = AutoProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")
```

```python
def img_caption(img_input: np.ndarray):
    # Converting numpy array to PIL Image and then to RGB image
    rgb_image = Image.fromarray(img_input).convert('RGB')
    # Processing
    inputs = processor(images=rgb_image, text="It is the image of", return_tensors="pt")
    # Generating a caption
    outputs = model.generate(**inputs, max_length=50)
    # Decoding the generated tokens to text
    caption = processor.decode(outputs[0], skip_special_tokens=True)
    return caption
```

```python
interface = gr.Interface(fn=img_caption, inputs=gr.Image(), outputs="text",
    title="Image Caption Generator",
    description="Image caption generator using a trained model."
)
interface.launch()
```
